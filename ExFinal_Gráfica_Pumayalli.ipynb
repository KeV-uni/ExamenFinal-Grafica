Respuesta:

- Un buffer es una memoria temporal que se utiliza para guardar datos durante un proceso en específico.

- Si no existiera un buffer, los datos se almacenarian en la memoria RAM, lo cual causaria que se llenara rapidamente de datos basura, que probablemte no vayamos a utilizar otra vez.

¿En qué se diferencia un espacio latente de un VAE y uno en una SR-GAN?

Respuesta:

El espacio latente es un espacio de representacion abstracta y de dimensionalidad reducida, en la cual se mapean datos de entrada en un vector de codificación, en donde cada dimensión representa algún atributo aprendido sobre los datos.

image.png

Por ejemplo para la imagen de arriba, un posible espacio latente contendria el siguiente vector de codificacion:

[Sonrisa: 0.99 \\
Tono de piel: 0.85 \\
Genero: -0.73 \\
Barba: 0.85 \\
Color de cabello: 0.68]

- La principal diferencia es que en el VAE (Varational AutoEnconder) el epsacio latente representa una distribución probabilística de las características latentes aprendidas por el modelo. Mientras que el espacion latente en SR-GAN se refiere al espacio de características de nivel superior aprendidas por el generador para y con el fin de mejorar la resolución de las imágenes
